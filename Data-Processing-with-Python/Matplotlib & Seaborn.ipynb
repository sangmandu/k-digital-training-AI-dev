### 2차함수 그래프 with plot()

```python
#np.arange(a,b,c) c: 0.01

x=np.arange(-10,10,0.01)
plt.plot(x, x**2)
plt.show()
```

```python
# x, y축에 설명 추가하기

x=np.arange(-10,10,0.01)

plt.xlabel("x value") ##
plt.ylabel("f(x) value") ##

plt.plot(x, x**2)

plt.show()
```

```python
# x, y축의 범위를 설정하기

x=np.arange(-10,10,0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.axis([-5, 5, 0, 25]) #[x_min, x_max, y_min, y_max]

plt.plot(x, x**2)

plt.show()
```

```python
# x, y축에 눈금 설정하기

x=np.arange(-10,10,0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.axis([-5, 5, 0, 25]) #[x_min, x_max, y_min, y_max]

plt.xticks([x for x in range(-5, 6)]) #x축의 눈금 설정
plt.yticks([x for x in range(0, 30, 5)]) #y축의 눈금 설정

plt.plot(x, x**2)

plt.show()
```

```python
# 그래프의 title 달기

x=np.arange(-10,10,0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.axis([-5, 5, 0, 25]) #[x_min, x_max, y_min, y_max]
plt.xticks([x for x in range(-5, 6)]) #x축의 눈금 설정
plt.yticks([x for x in range(0, 30, 5)]) #y축의 눈금 설정

plt.title("f(x)=x^2 graph") #그래프의 tilte 설정

plt.plot(x, x**2, label="trend") #legend를 그리는건 이 이후
plt.legend() #범례

plt.show()
```

## III. Matplotlib Case Study

+ 꺾은선 그래프 (Plot) : `plt.plot()`
+ 산점도 (Scatter Plot) : `plt.scatter()`
+ 박스그림 (Box Plot) : `plt.boxplot()`
+ 막대그래프 (Bar Chart) : `plt.bar()`
+ 원형그래프 (Pie Chart) : `plt.pie()`

### 꺾은선 그래프 (Plot)

+ `plt.plot()`

```python
x=np.arange(20) #0~19
y=np.random.randint(0, 20, 20) #난수를 20번 생성


# y축을 5단위로 20까지 보이게 하고 싶다면?
plt.axis([0, 20, 0, 20]) #[x_min, x_max, y_min, y_max]
plt.yticks([x for x in range(0,25,5)]) #[0,5,10,15,20]

plt.plot(x,y)
plt.show()

#시계열 데이터에서 많이 활용
```

### 산점도 (Scatter Plot)

+ `plt.scatter()`

```python
plt.scatter(x,y)
plt.show()

#x와 y가 별개인 변수일때, 산점도를 통해 상관관계 유추
```

### 박스 그림 (Box Plot)

+ 수치형 데이터에 대한 정보 (Q1, Q2, Q3, min, max)
+ `plt.boxplot()`

```python
plt.boxplot((x,y)) #두개의 변수의 분포를 비교할 수 있음

#plt의 title을 "Box plot of x, y"
plt.title("Box plot of x, y")
plt.show()

#데이터의 분포를 한번에 보기 쉬움
```

### 막대 그래프 (Bar Plot)

+ 범주형 데이터의 "값"과 그 값의 크기를 직사각형으로 나타낸 그림
+ `plt.bar()`

```python
plt.bar(x,y)

#xticks를 올바르게 처리
plt.xticks([i for i in x]) #np.arange(0,20,1)

plt.show()
```

```python
# cf) Histogram
# 하스토그램,막대그래프(같은점) : 도수분포를 직사각형의 막대 형태로 나타냄
# 히스토그램, 막대그래프(다른점) : 히스토그램은 여러 변량을 묶은 "계급"으로 표시
# 즉, "범주형" 데이터로 구성 후 그림.

plt.hist(y, bins=np.arange(0,22,2)) #bins : 기준
plt.xticks(np.arange(0,22,2)) #올바르게 xticks 고치기

plt.show()

# Histogram-막대간의 간격이 없음(이어져있음),보통 범주가 연속성을 가진경우가 많아서.
# 막대그래프-반대로 변량이 연속적이지 않은 경우가 있기때문에 띄어서 표시함
```

### 원형 그래프 (Pie Chart)

+ 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프
+ 다른 그래프에 비해 **비율** 확인에 용이
+ `plt.pie()`

```python
z=[100,300,200,400]

plt.pie(z, labels=["one", "two", "three", "four"]) #label 붙여주는게 보기 편함
plt.show()
```

## IV. Seaborn Case Study

+ 커널밀도그림 (Kernel Density Plot)

✨ [참고하기](https://seaborn.pydata.org/tutorial.html)

### Seabon Import 하기

```python
import seaborn as sns
```

### 커널밀도그림 (Kernel Density Plot)

+ 히스토그램과 같은 연속적인 분포를 곡선화해서 그린 그림
+ `sns.kdeplot()`

```python
# in Historgram

x=np.arange(0,22,2)
y=np.random.randint(0,20,20)

plt.hist(y, bins=x)
plt.show()
```

```python
# kdeplot : 어떤 데이터의 분포를 pdf처럼 확인할수있음 

sns.kdeplot(y, shade=True) #shade ; 그래프 아래있는 부분을 색칠할건지 여부 (boolean)

plt.show() #seaborn에서 plt를 기반으로해서 기본적으로 plt.imshow()와 같은 명령을 공유
```

### 카운트그림 (Count Plot)

+ 범주형 Ccolumn의 빈도수를 시각화 -> Groupby 후의 도수를 하는 것과 동일한 효과
+ `sns.countplot()`

```python
vote_df=pd.DataFrame({"name":["Andy", "Bob", "Cat"], "vote":[True, True, False]})

vote_df
```

```python
# in matplolib barplot

vote_count=vote_df.groupby("vote").count()

vote_count
```

```python
plt.bar(x=[False, True], height=vote_count["name"])

plt.show()
```

```python
# sns의 countplot

sns.countplot(x=vote_df['vote']) #해당 column

plt.show()
```

### 캣그림 (Cat Plot)

+ 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수
+ 복잡한 데이터 다룰 때 사용
+ `sns.catplot()`

```python
covid=pd.read_csv("./country_wise_latest.csv")

covid.head(5)
```

```python
s = sns.catplot(x="WHO Region", y="Confirmed", data=covid, kind="violin") 
#default kind="strip"

s.fig.set_size_inches(10,6) #sns에서 figsize 조정

plt.show()
```

### 스트립그림 (Strip Plot)

+ scatter plot과 유사하게 데이터의 수치를 표현하는 그래프
+ `sns.strip()`

```python
sns.stripplot(x="WHO Region", y="Recovered", data=covid) #지역별로 recover되었는지

plt.show()
```

```python
# cf) swarmplot

sns.swarmplot(x="WHO Region", y="Recovered", data=covid)

plt.show()
```

### 히트맵 (Heatmap)

+ 데이터의 행렬을 색상으로 표현해주는 그래프
+ `sns.heatmap()`

```python
# covid 데이터에서의 상관 계수
covid.corr()
```

```python
sns.heatmap(covid.corr()) #상관관계를 히트맵으로

plt.show() #1에 가까우면(밝은 색) : 양의 상관관계, 0에 가까우면(어두운색) : 상관관계 x
```
